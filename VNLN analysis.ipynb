## Combine posts

```python
def video_dict(vid1, vid2, vid3):
    
    '''This function combines information for each
        video from TikTok and gives a dataframe as the 
        output.'''
    
    # video description
    description = [vid1['text'], vid2['text'], vid3['text']]
    # video length
    duration = [vid1['videoMeta']['duration'], vid2['videoMeta']
                ['duration'], vid3['videoMeta']['duration']]
    # timestamp
    createTime = [vid1['createTime'], vid2['createTime'], vid3['createTime']]
    # no. plays
    playCount = [vid1['playCount'], vid2['playCount'], vid3['playCount']]
    # no. comments
    commentCount = [vid1['commentCount'],
                    vid2['commentCount'], vid3['commentCount']]
    # no. shares
    shareCount = [vid1['shareCount'], vid2['shareCount'], vid3['shareCount']]
    # no. hearts
    diggCount = [vid1['diggCount'], vid2['diggCount'], vid3['diggCount']]
    # no. mentions
    mentions = [vid1['mentions'], vid2['mentions'], vid3['mentions']]

    lists = [description, duration, createTime, playCount,
        commentCount, shareCount, diggCount, mentions]
    
    df = pd.DataFrame(lists).transpose()
    df.columns = ['description', 'duration', 'createTime', 'playCount',
        'commentCount', 'shareCount', 'diggCount', 'mentions']
    
    return df
```
